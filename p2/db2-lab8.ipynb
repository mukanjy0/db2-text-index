```python
# %pip install matplotlib
import matplotlib.pyplot as plt
```

```python
def readData(filename: str):
  x = []
  y1 = []
  y2 = []
  with open(filename) as file:
    [xTitle, y1Title, y2Title] = file.readline().strip().split(sep=' ')

    for line in file:
      [cx, cy1, cy2] = line.strip().split(sep=' ')
      x.append(int(cx))
      y1.append(float(cy1))
      y2.append(float(cy2))
  return xTitle, x, y1Title, y1, y2Title, y2
```

```python
[xTitle, x, y1Title, y1, y2Title, y2] = readData("data.txt")
```

```python
fig, ax = plt.subplots()

ax.set_xscale('log')

ax.plot(x, y1, 'go-', label=y1Title)
ax.plot(x, y2, 'o-', color='orange', label=y2Title)

ax.set_xlabel("DB size")
ax.set_ylabel("Query time")
ax.set_title("DB size vs query time with & without index")

ax.legend()

plt.savefig('fig.png', dpi=300)
plt.show()
```
