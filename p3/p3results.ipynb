## Gr√°ficas de resultados

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

# Leer los datos del archivo data.txt
sizes = []
no_index_means = []
index_means = []

with open('data.txt', mode='r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        sizes.append(int(row['Size']))
        no_index_means.append(float(row['NoIndex']))
        index_means.append(float(row['Index']))

# Convertimos los tama√±os a n√∫meros para el eje x
x = np.array(sizes)

# Creamos la figura y los ejes
plt.figure(figsize=(10, 6))
plt.plot(x, no_index_means, marker='o', label='Sin √≠ndice')
plt.plot(x, index_means, marker='o', label='Con √≠ndice')

# A√±adimos etiquetas y t√≠tulo
plt.xlabel('Tama√±o')
plt.ylabel('Tiempo promedio (ms)')
plt.title('Comparaci√≥n de tiempos con y sin √≠ndice')
plt.xscale('log')  # Usamos escala logar√≠tmica para el eje x
plt.yscale('log')  # Usamos escala logar√≠tmica para el eje y
plt.xticks(x, [str(size) for size in sizes])
plt.legend()

# Mostramos el gr√°fico
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
# Crear el segundo gr√°fico sin escala logar√≠tmica
plt.figure(figsize=(10, 6))
plt.plot(x, no_index_means, marker='o', label='Sin √≠ndice')
plt.plot(x, index_means, marker='o', label='Con √≠ndice')
plt.xlabel('Tama√±o')
plt.ylabel('Tiempo promedio (ms)')
plt.title('Comparaci√≥n de tiempos con y sin √≠ndice (Escala Lineal)')
plt.xticks(x, [str(size) for size in sizes])
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
